```python
# Pakcage import
from tkinter import filedialog as fd
from collections import defaultdict
import matplotlib.pyplot as plt
import csv
import pandas as pd
import folium
```

```python
# Load tower informaiton
inputFileTower = open('tower_list.csv', 'r')
tower = defaultdict(int)
line = inputFileTower.readline()
line = inputFileTower.readline()
while line:
    fields = line.split(",")
    location = (float(fields[1].strip("\'")),float(fields[2].strip("\'\\\n")))
    tower[location]=int(fields[0].strip("\'"))
    line = inputFileTower.readline()
#print(tower)
inputFileTower.close()
```

```python
list_of_files = ['process_2/24_25_jun_process_1.csv','process_2/26_30_jun_process_1.csv','process_2/01_07_jul_process_1.csv']
oneUser = pd.read_csv('process_2/22_23_jun_process_1.csv')
oneUser = oneUser.loc[oneUser['\'user_id\''] == 3]

for item in list_of_files:
    df = pd.read_csv(item)
    df = df.loc[df['\'user_id\''] == 2000]
    oneUser = pd.concat([oneUser,df])
oneUser
```

```python
tower_filtered = set(oneUser["\'tower\'"].unique())
tower_filtered
```

```python
lat = []
lon = []
name = []
for key,value in tower.items():
    if value in tower_filtered:
        name.append(value)
        lat.append(key[0])
        lon.append(key[1])

```

```python
# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lat':lat,'lon':lon,'name':name
}, dtype=str)

data
```

```python

# Make an empty map
m = folium.Map(location=[23.8139, 90.3986], tiles="OpenStreetMap", zoom_start=10)
```

```python
for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
      popup=data.iloc[i]['name'],
   ).add_to(m)

# Show the map again
m
```
